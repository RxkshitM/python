```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
flat_chess = cv2.imread('/content/chessboard.png')
```

```python
cv2_imshow(flat_chess)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
```

```python
gray = np.float32(gray_flat_chess)
```

```python
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
flat_chess[dst>0.01*dst.max()]= [255,0,0]
```

```python
plt.imshow(flat_chess)
```

```python
real_chess = cv2.imread('/content/chessBoardFull.png')
```

```python
plt.imshow(real_chess)
```

```python
real_chess_gray = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

```python
gray = np.float32(real_chess_gray)
```

```python
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
real_chess [dst>0.01*dst.max()]=[0,255,0]
```

```python
plt.imshow(real_chess)
```

# SHITO MASI CORNER

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,5,0.01,10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
plt.imshow(flat_chess)
```
